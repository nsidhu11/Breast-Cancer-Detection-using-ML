```python
# IMPORTING LIBRARIES

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Importing data
data = pd.read_csv('./data.csv')
```

```python
y=data.diagnosis
y[:25]

         # M = Malignant and B = Benign
```

```python
# Creating an array of attributes to be dropped
# Droping 'diagnosis' from input as it is target variable
# Droping 'id' and 'unnamed: 32' because they are not important

toBeDropped=['id','diagnosis','Unnamed: 32']
x=data.drop(toBeDropped,axis=1)      #[axis=0 for rown] and [axis=1 for columns]
x.head()
```

```python
#finding the value count of target variable
ax = sns.countplot(y, label="Count")
B, M = y.value_counts()
print(M, B)
```

```python
x.describe()
```

```python
# plotting correlation heatmap
f,ax = plt.subplots(figsize=(18,18))
sns.heatmap(x.corr(), annot=True, linewidths=.5, fmt= '.1f',ax=ax)
```

```python
#finding highly correlated features and substituting each group with one of the values

drop_list = ['perimeter_mean','radius_mean','compactness_mean','concave points_mean','radius_se','perimeter_se','radius_worst','perimeter_worst','compactness_worst','concave points_worst','compactness_se','concave points_se','texture_worst','area_worst']
x_1 = x.drop(drop_list, axis=1)

# Plotting the final heat-map
f,ax = plt.subplots(figsize=(14, 14))
sns.heatmap(x_1.corr(), annot=True, linewidths=.5, fmt= '.1f',ax=ax)
```

```python
# replacing M with 1 and B with 0
y_new = y.apply(lambda x: '1' if x == 'M' else '0')
```

```python
# importing libraries for training and testing the model

from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import f1_score,confusion_matrix
from sklearn.metrics import accuracy_score
```

```python
# splitting the data into training and testing
x_train,x_test,y_train,y_test=train_test_split(x_1,y_new,test_size=0.3,random_state=42)
```

```python
#initialize random forest classifier
RF=RandomForestClassifier()
```

```python
#train the classifier
RF=RF.fit(x_train,y_train)
```

```python
#testing
y_pred=RF.predict(x_test)
```

```python
#accuracy
acc=accuracy_score(y_test,y_pred)
acc
```
